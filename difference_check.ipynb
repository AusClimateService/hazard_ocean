```python
import xarray as xr
import matplotlib.pyplot as plt
```

```python
def load_result(file_path):
    return xr.open_dataset(file_path)
```

```python
def plot_variable(variable, title):
    plt.figure(figsize=(10, 6))
    variable.plot()
    plt.title(title)
    plt.show()
```

```python
def compare_results(gw1p2, dur_gw1p2_path, mag_gw1p2_path):
    dur_gw1p2 = xr.open_dataset(dur_gw1p2_path)
    mag_gw1p2 = xr.open_dataset(mag_gw1p2_path)

    # Calculate differences
    dur_diff = gw1p2['duration'] - dur_gw1p2['temp']
    mag_diff = gw1p2['magnitude'] - mag_gw1p2['temp']

    # Plot duration and magnitude
    plot_variable(gw1p2['duration'], 'Duration - GW1p2')
    plot_variable(gw1p2['magnitude'], 'Magnitude - GW1p2')

    # Plot scratch file duration and magnitude
    plot_variable(dur_gw1p2['temp'], 'Duration - scratch')
    plot_variable(mag_gw1p2['temp'], 'Magnitude - scratch')

    # Plot differences
    plot_variable(dur_diff, 'Difference in Duration (GW1p2 - scratch)')
    plot_variable(mag_diff, 'Difference in Magnitude (GW1p2 - scratch)')
```

```python
def main():
    # Load the GW1p2 results
    gw1p2_path = '/g/data/xv83/users/ep5799/Heatwaves/GW1p2_mhw.nc'
    gw1p2 = load_result(gw1p2_path)

    # Paths to scratch files
    dur_gw1p2_path = '/scratch/xv83/rxm599/dur_GW1p2.nc'
    mag_gw1p2_path = '/scratch/xv83/rxm599/mag_GW1p2.nc'

    # Compare the results
    compare_results(gw1p2, dur_gw1p2_path, mag_gw1p2_path)

if __name__ == "__main__":
    main()
```
